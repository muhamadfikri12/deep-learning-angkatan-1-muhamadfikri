# Definisi Deep Learning

Deep learning adalah bagian dari Machine Learning yang mengkhususkan diri dalam penggunaan jaringan saraf tiruan (Neaural Network) yang sangat dalam (dalam hal jumlah lapisan/layer yang dalam). Tujuannya untuk mempelajari representasi data yang semakin abstrak dan kompleks.

# Metode Deep Learning

## Feedforward Neural Network

Merupakan Jaringan Saraf Tiruan (Neural Network) yang terdiri dari lapisan-lapisan. Tujuan nya adalah untuk tugas seperti klasifikasi, regresi, dan pemrosesan data (data processing)

## Convolutional Neural Network (CNN)

CNN Biasanya digunakan untuk tugas-tugas pengolahan citra dan visual. Jadi metode ini menggunakan lapisan konvolusi untuk mengidentifikasi fiture-fiture spasial dalam citra.

## Generative Adversarial Network (GAN)

GAN memiliki dua jaringan, yaitu generator dan discriminator yang bersaing satu sama lain. Hal ini digunakan untuk menghasilkan data baru yang mirip dengan data yang ada, seperti generasi gambar.

## Autoencoders

Merupakan arsitektur yang digunakan untuk pembelajaran tanpa pengawasan dan reduksi dimensi data (Principal Component Analysis)

 ## Matplotlib

Tujuan : untuk membuat visualisasi statistik, interaktif, dan animasi.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace (0, 2*np.pi, 100)
y = np.sin (x)

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Grafik Sin')
plt.show()
```

## Scikit-Learn

Tujuan : untuk machine learning dan data mining

```python
from sklearn.linear_model import LinearRegression

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

#Membuat dan melatih model linear regression (dapat digunakan untuk memprediksi masa depan)
model = LinearRegression()
model.fit(x,y) #latihan datanya x hasilnya y

# Membuat prediksi
prediksi = model.predict([[4]])
print(prediksi)
```

## Tensorflow

Tujuan : untuk deep learning dan membangun Neural Network (Jaringan Syaraf)

```python
!pip install tensorflow
```

```python
import tensorflow as tf
import numpy as np

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation = 'relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer = 'adam', loss='mse')

#data
x = np.array ([1,2,3,4,5])
y = 2*x+1

#train modelnya
model.fit(x, y, epochs = 100)

#buat prediksi
prediksi = model.predict([4])
print(prediksi)
```
